```python
#Graph for Unknown Source Energy (keV) vs. time (minutes) used excel graph for curve fitting equation



import numpy as np               # We need to "load" any libraries we want to use.
import matplotlib as mpl
import matplotlib.pyplot as plt  # We create an object called "plt" that allows us to plot.


# Import curve fitting package from scipy
from scipy.optimize import curve_fit


# For 417.2 keV
x = np.array([0, 8, 18, 35, 61, 65, 76, 84, 96])
y1 = np.array([1350, 1245, 1052, 797, 628, 604, 452, 368, 289])

# y-uncertainty values
y1erra = np.array([3.32, 3.55, 3.98, 4.78, 5.51, 7.14, 8.90, 11.35, 14.08])

# Define our fit function
def fitfunc(x, a, b):
    return a*x + b

parsWithUnc, cov = curve_fit(f=fitfunc, xdata=x, ydata=y1, p0=[1,0], sigma=y1erra, absolute_sigma=True, bounds=(-2,20))
print("Fit Parameters (Uncertainty IS included):")
print("Slope:  a = ", parsWithUnc[0])
print("Offset: b = ", parsWithUnc[1])

#==========================================

# For 819.1 keV
y2 = np.array([338, 352, 339, 156, 157, 161, 136, 133, 127])

# y-uncertainty values
y2erra = np.array([9.15, 8.66, 8.10, 18.50, 15.30, 14.77, 18.87, 18.26, 17.97])

# Define our fit function
def fitfunc(x, a, b):
    return a*x + b

parsWithUnc, cov = curve_fit(f=fitfunc, xdata=x, ydata=y2, p0=[1,0], sigma=y2erra, absolute_sigma=True, bounds=(-2,20))
print("Fit Parameters (Uncertainty IS included):")
print("Slope:  a = ", parsWithUnc[0])
print("Offset: b = ", parsWithUnc[1])

#============================================

# For 1097.3 keV
y3 = np.array([1505, 1438, 1156, 817, 772, 601, 510, 599, 434])

# y-uncertainty values
y3erra = np.array([3.09, 3.19, 3.68, 4.87, 4.39, 5.65, 6.15, 5.10, 6.50])

# Define our fit function
def fitfunc(x, a, b):
    return a*x + b

parsWithUnc, cov = curve_fit(f=fitfunc, xdata=x, ydata=y3, p0=[1,0], sigma=y3erra, absolute_sigma=True, bounds=(-2,20))
print("Fit Parameters (Uncertainty IS included):")
print("Slope:  a = ", parsWithUnc[0])
print("Offset: b = ", parsWithUnc[1])

#============================================

# For 1293.6 keV
y4 = np.array([2064, 1816,1634, 1285, 962, 934, 760, 693, 593])

# y-uncertainty values
y4erra = np.array([2.36, 2.61, 2.65, 3.15, 3.91, 3.92, 4.50, 4.75, 5.48])

# Define our fit function
def fitfunc(x, a, b):
    return a*x + b

parsWithUnc, cov = curve_fit(f=fitfunc, xdata=x, ydata=y4, p0=[1,0], sigma=y4erra, absolute_sigma=True, bounds=(-2,20))
print("Fit Parameters (Uncertainty IS included):")
print("Slope:  a = ", parsWithUnc[0])
print("Offset: b = ", parsWithUnc[1])

#============================================

# Create the figure
fig, ax = plt.subplots()

# Create error bars. (You should try changing the settings here to see the effect)
ax.errorbar(x, y1, xerr=1, yerr=y1erra, fmt='bo', capsize=4, ecolor='red')
ax.errorbar(x, y2, xerr=1, yerr=y2erra, fmt='bo', capsize=4, ecolor='red')
ax.errorbar(x, y3, xerr=1, yerr=y3erra, fmt='bo', capsize=4, ecolor='red')
ax.errorbar(x, y4, xerr=1, yerr=y4erra, fmt='bo', capsize=4, ecolor='red')

ax.set_xlabel('Time in minutes')  # Add an x-label to the axes.
ax.set_ylabel('Counts without background')  # Add a y-label to the axes.

#plots of all 4 energies 
plt.semilogy(x,y1,linestyle='-', linewidth=2, color='black', label="417.2 keV")
plt.semilogy(x,y2,linestyle='-', linewidth=2, color='red', label="819.1 keV")
plt.semilogy(x,y3,linestyle='-', linewidth=2, color='blue', label="1097.3 keV")
plt.semilogy(x,y4,linestyle='-', linewidth=2, color='orange', label="1293.6 keV")

plt.legend()
plt.grid()
plt.show() 
```
